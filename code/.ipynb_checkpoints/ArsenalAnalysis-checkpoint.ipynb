```python
import pandas as pd
%matplotlib inline
```

```python
# Read the dataset of Premier League match results for the season 2018-2019 and store the results in a DataFrame
football_data = pd.read_csv('../data/EPL18_19.csv')
```

```python
# ARSENAL DATA ONLY
# Since, I am only concerned with 'Arsenal' matches, I choose only those games where 
# ('HomeTeam' == 'Arsenal') OR ('AwayTeam' == 'Arsenal')
arsenal_data = football_data[(football_data['HomeTeam'] == 'Arsenal') | (football_data['AwayTeam'] == 'Arsenal')]
```

```python
# Removing columns we do not require
arsenal_data = arsenal_data.loc[:,'Div':'AR']
arsenal_data.head()
```

```python
# Shape of the Arsenal data: Should be 38 - number of games in a single season.
total_games, total_features = arsenal_data.shape
print("The total number of games played by Arsenal both HOME and AWAY in 2018-2019 Premier League Season - {}"
     .format(total_games))
```

```python
home_data = arsenal_data[arsenal_data['HomeTeam'] == 'Arsenal']
home_data.head()
```

```python
home_games = home_data.shape[0]
print("The number of games played by Arsenal at HOME in 2018-2019 Premier League Season - {}"
     .format(home_games))
```

```python
# We can easily describe the statistics using .describe() method.
home_data.describe()
```

```python
# Average number of goals scored by Arsenal at HOME:
print("The average number of goals scored by Arsenal at HOME - {}".format(home_data['FTHG'].mean()))
```

```python
# Average number of goals conceded by Arsenal at HOME:
print("The average number of goals conceded by Arsenal at HOME - {}".format(home_data['FTAG'].mean()))
```

```python
# Total number of goals scored by Arsenal at HOME:
print("The total number of goals scored by Arsenal at HOME - {}".format(home_data['FTHG'].sum()))
```

```python
# Total number of goals conceded by Arsenal at HOME:
print("The total number of goals conceded by Arsenal at HOME - {}".format(home_data['FTAG'].sum()))
```

```python
# Average number of shots by Arsenal at HOME:
print("The average number of shots by Arsenal at HOME - {}".format(home_data['HS'].mean()))
```

```python
# Average number of shots conceded by Arsenal at HOME:
print("The average number of shots conceded by Arsenal at HOME - {}".format(home_data['AS'].mean()))
```
